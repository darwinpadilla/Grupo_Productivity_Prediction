```python
# DP GIT. 1.6 Conversion de variables categóricas
#Determinamos que columnas (variables) son del tipo númerico y cuales son categóricas para poder usar esta información más adelante:
numeric_columns = list()
categorical_columns = list()
dictionary_of_columns_with_index_to_categorical = dict()
dictionary_of_columns_with_categorical_to_index = dict()

for column in dataset:
  #Determinamos si la columna (variable) es numérica o no
  if pd.api.types.is_numeric_dtype(dataset[column]):
    numeric_columns.append(column)
  else:
    #Modificamos el tipo de dato de la columna (variable) mediante "astype"
    dataset[column] = dataset[column].astype("category")

    #Verificamos si el tipo de dato de la columna (variable) fue transformado a categórico correctamente
    if not pd.api.types.is_categorical_dtype(dataset[column]):
      raise Exception("La columna {} no se transformó correctamente a categórica".format(column))

    dictionary_of_columns_with_index_to_categorical[column] = dict()
    dictionary_of_columns_with_categorical_to_index[column] = dict()
    
    #Indexamos los valores (categorías), sin tomar en consideración los nulos, de la columna (variable) y guardamos esa información en los diccionarios
    for index, category in enumerate(dataset[column].cat.categories):
      dictionary_of_columns_with_index_to_categorical[column][index] = category
      dictionary_of_columns_with_categorical_to_index[column][category] = index
    
    categorical_columns.append(column)
```

**One-hot encoding**

Para ayudar a la predicción de productividad vamos aplicar One-hot en las siguientes variables:
1.   mes
2.   semana
3.   department
4.   dia_semana
5.   team



```python
# RD_NUEVO
def crear_one_hot_encoding(ds_para_encoding,nombre_columna):
    nuevas_col_encoding = pd.get_dummies(ds_para_encoding[nombre_columna],prefix=nombre_columna)
    ds_para_encoding = pd.concat([ds_para_encoding,nuevas_col_encoding],axis=1)
    return ds_para_encoding

ds_binario = crear_one_hot_encoding(ds_binario,"mes")
ds_binario = crear_one_hot_encoding(ds_binario,"semana")
ds_binario = crear_one_hot_encoding(ds_binario,"department")
ds_binario = crear_one_hot_encoding(ds_binario,"dia_semana")
ds_binario = crear_one_hot_encoding(ds_binario,"team")

ds_binario.columns
```

**Creando la variable a predecir "TARGET_BINARIO"**

```python
# RD_NUEVO
# se procede a hallar productividad_residual que es la productividad lograda menos la esperada
# si la productividad_residual es mayor igual a 0 estamos en un rendimiento esperado
# si la productividad_residual es menor a cero estamos en un bajo rendimiento
ds_binario['productividad_residual']=ds_binario.actual_productivity - ds_binario.targeted_productivity
ds_binario.columns
ds_binario.head()
```

```python
# RD_NUEVO
ds_binario['productividad_residual'].describe()
```

```python
# RD_NUEVO
ds_binario['TARGET_BINARIO_PRODUCTIVIDAD']=np.nan
ds_binario.loc[ds_binario['productividad_residual']<0,'TARGET_BINARIO_PRODUCTIVIDAD'] = -1
ds_binario.loc[(ds_binario['productividad_residual']==0), 'TARGET_BINARIO_PRODUCTIVIDAD'] = 0
ds_binario.loc[ds_binario['productividad_residual']>0, 'TARGET_BINARIO_PRODUCTIVIDAD'] = 1
ds_binario.head()
```

**ENTRENAMIENTO Y EVALUACION AL DATASET PARA CLASIFICACION BINARIA**

```python
# RD_NUEVO

definicion_metricas = pd.DataFrame(columns=[
    'Modelo', 'Precisión del Entrenamiento', 'Precisión de la Prueba', 'Precisión',
    'Recall','f1_score_test'
])

modelos_de_entrenamiento = [LogisticRegression(),
               DecisionTreeClassifier(),
               RandomForestClassifier(),
               SVC(),
               KNeighborsClassifier(),
               GaussianNB(),LinearDiscriminantAnalysis()
               ]

def obtener_performance_de_metricas(model, i):
    # model name
    nombre_de_modelo = type(model).__name__
    print("Entrenando modelo {} ...".format(nombre_de_modelo))
    # Fitting of model
    model.fit(ds_binario1_X_train, ds_binario1_y_train)
    print("Completado modelo {}  entrenado.".format(nombre_de_modelo))
    # Predictions
    y_pred = model.predict(ds_binario1_X_test)
    # Add to ith row of dataframe - metrics

    definicion_metricas.loc[i] = [
        nombre_de_modelo,
        model.score(ds_binario1_X_train, ds_binario1_y_train),
        model.score(ds_binario1_X_test, ds_binario1_y_test),
        precision_score(ds_binario1_y_test, y_pred),
        recall_score(ds_binario1_y_test, y_pred),
        f1_score(ds_binario1_y_test, y_pred),
    ]
   
    print("Completado Evaluación del Modelo {} .".format(nombre_de_modelo))
```

```python
# RD_NUEVO
from sklearn.metrics import r2_score,f1_score
for n, modelo_class in enumerate(modelos_de_entrenamiento):
    obtener_performance_de_metricas(modelo_class, n)
```

```python
# RD_NUEVO
def mostrar_resultados(resultados_class):
    print(f'Los mejores parametros son: {resultados_class.best_params_}')
    print("\n")
    mean_score_class = resultados_class.cv_results_['mean_test_score']
    std_score_class = resultados_class.cv_results_['std_test_score']
    params_class = resultados_class.cv_results_['params']
    for mean,std,params in zip(mean_score_class,std_score_class,params_class):
        print(f'{round(mean,3)} + or -{round(std,3)} for the {params}')
```

```python
# RD_NUEVO
mostrar_resultados(cv_class)
```
